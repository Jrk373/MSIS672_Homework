# Missing Data

## Impute values

```python
# replace the missing values using the median of the remaining values.
#calculate median value 
medianAvg_Cost_Emergency_Dept_Services = CHRoster_df['Avg_Cost_Emergency_Dept_Services'].median() 
medianAvg_Cost_Emergency_Dept_Services
```

```python
# replace the missing values with the median value
CHRoster_df.Avg_Cost_Emergency_Dept_Services = CHRoster_df.Avg_Cost_Emergency_Dept_Services.fillna(value=medianAvg_Cost_Emergency_Dept_Services) 

print('Number of rows with valid Avg_Cost_Emergency_Dept_Services values after filling NA values: ', CHRoster_df['Avg_Cost_Emergency_Dept_Services'].count())
```
